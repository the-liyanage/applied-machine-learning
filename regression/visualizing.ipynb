### Box plots

1) where is the median line sitting? ---> tells you the typical value for that group. compare across groups to see which one generally performs higher or lower.
   
2) how big is the box? ---> bigger box means high variability, data is spread out. smaller box means most values are clustered together and consistent.

3) how long are the whiskers? ---> longer whiskers mean the data stretches far beyond the middle 50%. short whiskers mean even the extremes are close to the center.

4) are there dots beyond the whiskers? ---> yes dots means unusual values exist. worth investigating why. no dots means the data is stable and predictable.

   
and if you're comparing multiple box plots also ask — do the boxes overlap? if yes the groups are similar. if one box sits completely above another the groups are genuinely different.

### Bar charts


1) which bar is the tallest? ---> that's the most frequent or highest value category in your data.
   
2) which bar is the shortest? ---> least frequent or lowest. if it's very small compared to others, that group is underrepresented and conclusions about it should be made carefully.

3) are the bars roughly equal in height? ---> if yes your data is balanced across categories. if some bars dominate, your data is imbalanced.

4) is there a clear pattern or ranking? ---> bars steadily decreasing or increasing tells a story about order or priority in your data.

-------------------------------------------------------------------------------------------------------------------------------------------- 

### Pie charts

1) which slice dominates? ---> anything above 50% means one category is controlling the majority of your data.

2) are there very tiny slices? ---> small slices mean that category barely exists in your data. same as short bars — be cautious drawing conclusions about them.

3) is the distribution even or uneven? ---> roughly equal slices means balance across categories. heavily uneven slices means a few categories dominate.

4) do a few categories together make up most of the pie? ---> if two or three slices together cover 80%+ of the pie, the remaining categories are minor and may not need much attention in your analysis.

- ------------------------------------------------------------------------------------------------------------------------------------------

### Violin plots


1) where is the violin widest? ---> that's where most of your data lives.
   
2) does it have one bulge or two? ---> one bulge means your population is uniform, two bulges means there are two distinct groups worth investigating.
   
3) how long and thin is the tail? ---> longer thinner tail means a few extreme values exist but they're rare.

4) do multiple violins look similar in shape? ---> if yes, those variables are likely correlated and worth checking with a correlation matrix.

   
those four questions work whether you're looking at exam scores, house prices, customer ages or anything else.

### Every Categorical Variable 
### 1. Box plot Comparisons

for each group, 
1) Which group has a higher median?  -> who performs better on average
2) Are the IQRs (box sizes) different? -> who is more consisten
3) Which group has more/worse outliers? -> who has the struggling tail
4) Do the boxes OVERLAP significantly? -> is the difference meaningful
