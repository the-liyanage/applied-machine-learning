### 1. Missing Values 

```python
#We will check the percentage of NaN values present in each feature
print("All the features which has missing NaN values:\n")
features_with_na = [features for features in df.columns if df[features].isnull().sum()>1]
print(features_with_na)
print()

print(f"{'Feature':<30} {'Missing Values(mean)':<50}")
for feature in features_with_na:
    print(f"{feature:<30} {np.round(df[feature].isnull().mean(), 4)}")
    
```

### 1.1 Since there are many missing values, we need to find the relationship between missing values and the Sales Price

```python
for feature in features_with_na:
    data = df.copy()

    #Creating a varibale indicates 1 --> if the observation was missing, 0 --> if not missing
    data[feature] = np.where(data[feature].isnull(), 1,0)

    #Let's calculate the mean SalesPrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar(color=['skyblue', 'coral'])
    plt.title(feature)
    plt.show()
    
```

- Here with the relationship between the missing values and the dependent variable is clearly visible.

- We need to replace these NaN values with something meaningful, which we will do in the Feature Engineering section.

### Temporal Variables

- Temporal variables are variables that relate to time or have a time-based component.
- They capture when something happened, how long it lasted, or patterns over time.

- Fromt he Dataset we have 4 variables
     - YearBuilt
     - YearRemodAdd
     - GarageYrBlt
     - YrSold

```python
#list of variables that contain temporal variables
temporal_features = [features for features in df.columns if 'Yr' in features or 'Year' in features]
print(temporal_features)
```

```python
#Unique content of these features

for feature in temporal_features:
    print(feature, df[feature].unique())
```

```python
#let's analyze the relation between temporal variables and SalesPrice
plt.figure(figsize = (10,6))
df.groupby("YrSold")["SalePrice"].median().plot()
plt.xlabel("Year Sold")
plt.ylabel("Median House Price")
plt.title("House Prices Vs Year Sold")
```

```python
#We will compare the dfferences between ALL year features with SalesPrice

for feature in temporal_features:
    if feature != 'YrSold':
        data = df.copy()
        #We will capture differences between YrSold feature and other temporal features

        data[feature] = data["YrSold"] - data[feature]
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("Sales Price")
        plt.show()


```
