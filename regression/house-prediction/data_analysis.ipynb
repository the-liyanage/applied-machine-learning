### 1. Missing Values 

```python
#We will check the percentage of NaN values present in each feature
print("All the features which has missing NaN values:\n")
features_with_na = [features for features in df.columns if df[features].isnull().sum()>1]
print(features_with_na)
print()

print(f"{'Feature':<30} {'Missing Values(mean)':<50}")
for feature in features_with_na:
    print(f"{feature:<30} {np.round(df[feature].isnull().mean(), 4)}")
    
```

### 1.1 Since there are many missing values, we need to find the relationship between missing values and the Sales Price

```python
for feature in features_with_na:
    data = df.copy()

    #Creating a varibale indicates 1 --> if the observation was missing, 0 --> if not missing
    data[feature] = np.where(data[feature].isnull(), 1,0)

    #Let's calculate the mean SalesPrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar(color=['skyblue', 'coral'])
    plt.title(feature)
    plt.show()
    
```
