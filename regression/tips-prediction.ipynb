### Waiter Tips Prediction with Machine Learning



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/amankharwal/Website-data/master/tips.csv"
df = pd.read_csv(url)

df.to_csv("tips.csv", index = False)
df = pd.read_csv("tips.csv")
print(f"Shape of the Dataset:\n {df.shape}\n")
print(df.head())
```

### 1.0 Data Cleaning 

```python
#Checking for the null values
print(f"{df.isnull().sum()}\n")

print(f"{df.info()}\n")
```

```python
#Checking for the duplicated values
print(df.duplicated())
```

```python
#Identifying the Column data types

cat_col = [feature for feature in df.columns if df[feature].dtype == "object"]
num_col = [feature for feature in df.columns if df[feature].dtype != "object"]

print(f"Categorical columns:\n{cat_col}\n")
print(f"Numerical columns:\n{num_col}\n")
```

```python
#Counting the Unique values in Categorical Columns
print(df[cat_col].nunique())
```

### 2.0 Visualzation of Data 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = df,
                x = "total_bill",
                y = "tip",
                hue = "day", #Color by day
                size = "size", #Size by the number of people 
                sizes = (30, 300),
                alpha = 0.7 )
plt.grid(True, alpha = 0.3)
plt.title("Tips Vs Total Bill by Day and Number of People")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = df,
                x = "total_bill",
                y = "tip",
                hue = "sex",
                size = "size",
                sizes = (30, 300),
                alpha = 0.7 )
plt.grid(True, alpha = 0.3)
plt.title("Tips Vs Total Bill by Gender of Person")
plt.show()
```
