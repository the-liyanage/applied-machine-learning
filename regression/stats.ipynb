### Median

```
Line up all your female students by their math score, shortest to tallest. The median is the girl standing exactly in the middle.

if median = 65,
your typical female student scores 65 in math

```

### The Median Gap

```
Female Math Median: 65
Male Math Median:   69
Gap:                4 points

If you picked one random female and one random male, the typical male scores
about 4 points higher in math

not evert make beats every female. Just the typical one does.

```

### IQR - How different are students from each other 

```
You have two classes
Class A scores: 60, 62, 63, 65, 65, 66, 67, 68, 70
Class B scores: 20, 35, 45, 65, 65, 75, 85, 90, 100

Both classes have the same median(65), But Class B is all over the place

IQR measures that 'all over the place' feeling
Class A IQR: small  → students are SIMILAR to each other
Class B IQR: large  → students are VERY DIFFERENT from each other

IQR = how consistent is this group?
Small IQR --> students cluster together, predictable
Large IQR --> students are scattered, unpredictable

```
